```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
Omega_arr = np.array([0, 0.3, 0.7, 1])
z_arr = np.linspace(0, 1, 100)
for Omega in Omega_arr:
    result_arr = []
    for z in z_arr:
        result = integrate.quad(lambda z: 1/(Omega*(1+z)**3+(1-Omega)*(1+z)**(3/2))**(1/2), 0, z)
        if z == 0:
            result_arr = [result]
        else:
            result_arr.append(result)
    result_arr = np.array(result_arr)
    plt.plot(z_arr, result_arr[:,0], label=r'$\Omega$ = ' + str(Omega))
    
    
plt.plot(z_arr, 4*(1+z_arr)**(1/4) - 4, '--', label=r'$\Omega$ = 0 (Analytic Soln.)', c='orange', alpha=1)
plt.plot(z_arr, -2*(1+z_arr)**(-1/2) + 2, '--', label=r'$\Omega$ = 1 (Analytic Soln.)', c='yellow', alpha=1)
plt.xlabel('Redshift z')
plt.ylabel('Evaluation of Integral')
plt.title('Numerical Integration of Integral' + r'$ \int_0^z \frac{dz}{\left[\Omega (1+z)^3 + (1-\Omega) (1+z)^{3/2}\right]^{1/2}}$')
plt.legend()
plt.savefig('HW_1')
plt.show()
```
